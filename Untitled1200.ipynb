```python
# ==============================================
# SkillTracer — Recovery cell (fixed grouping dtype)
# - Normalizes student key to string to avoid KeyError in groupby.get_group
# - Loads/repairs model (.keras preferred, H5 fallback) and regenerates metrics/plots
# ==============================================
import os, csv, json, pickle, warnings, random
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

try:
    import seaborn as sns
    USE_SNS = True
except Exception:
    USE_SNS = False

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, confusion_matrix,
    roc_curve, auc, precision_recall_curve, average_precision_score,
    f1_score, brier_score_loss
)

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# --------------------------
# Paths
# --------------------------
DATA_PATH = r"C:\Users\sagni\Downloads\SkillTracer Knowledge Tracing\archive (1)\2012-2013-data-with-predictions-4-final.csv"
OUTPUT_DIR = r"C:\Users\sagni\Downloads\SkillTracer Knowledge Tracing"
os.makedirs(OUTPUT_DIR, exist_ok=True)

KERAS_PATH = os.path.join(OUTPUT_DIR, "model.keras")
H5_PATH    = os.path.join(OUTPUT_DIR, "model.h5")
PP_PATH    = os.path.join(OUTPUT_DIR, "preprocessor.pkl")

# --------------------------
# Reproducibility
# --------------------------
SEED = 42
np.random.seed(SEED); random.seed(SEED); tf.random.set_seed(SEED)

# --------------------------
# Utils
# --------------------------
def is_zip_or_xlsx(path):
    try:
        with open(path, "rb") as f:
            return f.read(2) == b"PK"
    except Exception:
        return False

def robust_read_any(path):
    if not os.path.exists(path):
        raise FileNotFoundError(path)
    if is_zip_or_xlsx(path):
        import openpyxl
        df = pd.read_excel(path, engine="openpyxl")
        print(f"[INFO] Loaded as Excel: {os.path.basename(path)} shape={df.shape}")
        return df
    encodings = ["utf-8","utf-8-sig","cp1252","latin1"]
    delimiters = [";", ",", "\t", "|"]
    try:
        with open(path, "rb") as f:
            head = f.read(8192).decode("latin1", errors="ignore")
        try:
            sniffed = csv.Sniffer().sniff(head)
            if sniffed.delimiter in delimiters:
                delimiters = [sniffed.delimiter] + [d for d in delimiters if d != sniffed.delimiter]
        except Exception:
            pass
    except Exception:
        pass
    last_err = None
    for enc in encodings:
        for sep in delimiters:
            try:
                df = pd.read_csv(path, encoding=enc, sep=sep, engine="python")
                if df.shape[1] > 1:
                    print(f"[INFO] Loaded as CSV enc='{enc}', sep='{sep}', shape={df.shape}")
                    return df
            except Exception as e:
                last_err = e
                continue
    raise RuntimeError(f"Could not parse {path}. Last error: {last_err}")

def pick_col(candidates, cols):
    for c in candidates:
        if c in cols: return c
    lc = {c.lower(): c for c in cols}
    for c in candidates:
        if c.lower() in lc: return lc[c.lower()]
    return None

# --------------------------
# Load dataset & columns
# --------------------------
df = robust_read_any(DATA_PATH)
cols = list(df.columns)
student_col = pick_col(["student_id","user_id","Anon Student Id","Anon StudentID","student","sid"], cols)
skill_col   = pick_col(["skill_id","skill","tag","KC(SubSkills)","KC","skill_name","concept_id"], cols)
correct_col = pick_col(["correct","is_correct","Correct First Attempt","answered_correctly","label"], cols)
time_col    = pick_col(["timestamp","start_time","order_id","time","Time"], cols)
if student_col is None or skill_col is None or correct_col is None:
    raise ValueError(f"Missing key columns in {cols}")

# Normalize correctness to {0,1}
df[correct_col] = pd.to_numeric(df[correct_col], errors="coerce")
df[correct_col] = (df[correct_col] > 0).astype(int)

# If multiple skills in one cell, take first
def take_first_skill(v):
    if pd.isna(v): return np.nan
    s = str(v)
    for sep in ["~~", "; ", ";", ",", "|"]:
        if sep in s: return s.split(sep)[0]
    return s

df[skill_col] = df[skill_col].apply(take_first_skill)

# Sort per student by time if available
if time_col is not None:
    try:
        df[time_col] = pd.to_datetime(df[time_col], errors="coerce")
        df = df.sort_values([student_col, time_col]).reset_index(drop=True)
    except Exception:
        df = df.sort_values([student_col]).reset_index(drop=True)
else:
    df = df.sort_values([student_col]).reset_index(drop=True)

# Drop missing essentials
df = df.dropna(subset=[student_col, skill_col, correct_col])

# --------------------------
# Load/build preproc (skills)
# --------------------------
if os.path.exists(PP_PATH):
    with open(PP_PATH, "rb") as f:
        preproc = pickle.load(f)
    skill2idx = preproc["skill2idx"]
    n_skills  = preproc["n_skills"]
    MAX_LEN   = preproc["max_len"]
    TOKEN_PAD = preproc["token_pad_id"]
    print("[INFO] Loaded preprocessor.pkl")
else:
    skills = df[skill_col].astype(str).unique().tolist()
    skill2idx = {s:i for i,s in enumerate(skills)}
    n_skills  = len(skill2idx)
    MAX_LEN   = 200
    TOKEN_PAD = 0
    preproc = {
        "skill2idx": skill2idx,
        "idx2skill": {i:s for s,i in skill2idx.items()},
        "n_skills":  n_skills,
        "max_len":   MAX_LEN,
        "token_pad_id": TOKEN_PAD,
        "token_vocab_size": 2*n_skills + 1,
        "token_rule": "token = 1 + skill_id + correctness * n_skills",
        "columns": {"student": student_col, "skill": skill_col, "correct": correct_col, "time": time_col}
    }
    with open(PP_PATH, "wb") as f:
        pickle.dump(preproc, f)
    print("[INFO] Built & saved preprocessor.pkl from data")

# --------------------------
# Tokenization helpers
# --------------------------
def to_token(skill_name:str, corr:int):
    if skill_name not in skill2idx: return None
    sid = skill2idx[skill_name]
    return 1 + sid + int(corr)*n_skills  # 1..2*n_skills (0 is PAD)

def make_sequences_for_student(sub):
    skills = sub[skill_col].astype(str).tolist()
    corr   = sub[correct_col].astype(int).tolist()
    toks, corr_clean = [], []
    for s, c in zip(skills, corr):
        t = to_token(s, c)
        if t is not None:
            toks.append(t); corr_clean.append(c)
    if len(toks) < 2: return []
    tokens = np.array(toks[:-1], dtype=np.int32)
    labels = np.array(corr_clean[1:], dtype=np.int32)
    chunks = []
    for start in range(0, len(tokens), MAX_LEN):
        xt = tokens[start:start+MAX_LEN]
        yt = labels[start:start+MAX_LEN]
        if len(xt) == 0: continue
        pad = MAX_LEN - len(xt)
        xt = np.pad(xt, (0,pad), constant_values=TOKEN_PAD)
        yt = np.pad(yt, (0,pad), constant_values=-1).astype(int)
        mask = (yt != -1).astype(np.float32)
        yt = np.where(yt==-1, 0, yt).astype(np.float32)
        chunks.append((xt, yt[:,None], mask))
    return chunks

# --------------------------
# >>> FIX: normalize student key to STRING for both split and grouping
# --------------------------
df["_sid_key"] = df[student_col].astype(str)

# Split by student (strings)
sid_all = df["_sid_key"].unique()
train_sids, test_sids = train_test_split(sid_all, test_size=0.15, random_state=SEED)
train_sids, val_sids  = train_test_split(train_sids, test_size=0.1765, random_state=SEED)

# Group by the SAME normalized key (strings)
g = df.groupby("_sid_key", sort=False)

def collect(sids):
    X, Y, W = [], [], []
    for sid in sids:
        if sid not in g.groups:   # safety against any stray id
            continue
        sub = g.get_group(sid)
        for xt, yt, mw in make_sequences_for_student(sub):
            X.append(xt); Y.append(yt); W.append(mw)
    if len(X)==0: return None, None, None
    return np.stack(X), np.stack(Y), np.stack(W)

# Build splits
X_train, Y_train, W_train = collect(train_sids)
X_val,   Y_val,   W_val   = collect(val_sids)
X_test,  Y_test,  W_test  = collect(test_sids)
if X_test is None:
    raise RuntimeError("Empty test split after filtering unknown skills.")
print(f"[INFO] X_train={None if X_train is None else X_train.shape}, X_val={None if X_val is None else X_val.shape}, X_test={X_test.shape}")

# --------------------------
# Model build / load / recovery
# --------------------------
VOCAB_SIZE   = 2*n_skills + 1
EMBED_DIM    = 64
HIDDEN_UNITS = 128

def build_model():
    inp = keras.Input(shape=(MAX_LEN,), dtype="int32")
    x = layers.Embedding(input_dim=VOCAB_SIZE, output_dim=EMBED_DIM, mask_zero=True)(inp)
    x = layers.LSTM(HIDDEN_UNITS, return_sequences=True)(x)
    x = layers.Dropout(0.2)(x)
    out = layers.Dense(1, activation="sigmoid")(x)
    m = keras.Model(inp, out)
    m.compile(optimizer="adam", loss="binary_crossentropy", metrics=[keras.metrics.BinaryAccuracy(name="accuracy")])
    return m

model = None
loaded = False

# Prefer native .keras
if os.path.exists(KERAS_PATH):
    try:
        model = tf.keras.models.load_model(KERAS_PATH, safe_mode=False)
        loaded = True
        print(f"[INFO] Loaded native model: {KERAS_PATH}")
    except Exception as e:
        print("[WARN] Failed to load model.keras:", e)

# Legacy H5 fallback
if not loaded and os.path.exists(H5_PATH):
    try:
        model = tf.keras.models.load_model(H5_PATH, compile=False)
        loaded = True
        print(f"[INFO] Loaded legacy H5 model: {H5_PATH}")
    except Exception as e:
        print("[WARN] Could not deserialize H5 model:", e)
        model = build_model()
        try:
            model.load_weights(H5_PATH)
            loaded = True
            print("[INFO] Loaded weights from model.h5 into rebuilt architecture.")
        except Exception as e2:
            print("[WARN] Could not load weights from H5:", e2)

# Final fallback: brief retrain (if needed)
if not loaded:
    if X_train is None or X_val is None:
        raise RuntimeError("Cannot retrain fallback; training splits are empty.")
    print("[INFO] Retraining briefly to regenerate a valid model...")
    model = build_model()
    early = keras.callbacks.EarlyStopping(monitor="val_accuracy", mode="max", patience=3, restore_best_weights=True)
    model.fit(
        X_train, Y_train,
        sample_weight=W_train,
        validation_data=(X_val, Y_val, W_val),
        epochs=10, batch_size=64, callbacks=[early], verbose=1
    )

# Save native model for future use
try:
    model.save(KERAS_PATH)
    print(f"[INFO] Saved native Keras model -> {KERAS_PATH}")
except Exception as e:
    print("[WARN] Could not save model.keras:", e)

# --------------------------
# Evaluate on test and save plots/metrics
# --------------------------
y_prob_seq = model.predict(X_test, batch_size=128).squeeze(-1)  # (N, T)
mask = W_test.astype(bool)
y_true = Y_test.squeeze(-1)[mask]
y_prob = y_prob_seq[mask]

# Metrics @ 0.5
from sklearn.metrics import classification_report
y_pred_05 = (y_prob >= 0.5).astype(int)
acc_05 = accuracy_score(y_true, y_pred_05)
fpr, tpr, _ = roc_curve(y_true, y_prob); roc_auc = auc(fpr, tpr)
precision, recall, _ = precision_recall_curve(y_true, y_prob); ap = average_precision_score(y_true, y_prob)
brier = brier_score_loss(y_true, y_prob)
print(f"\n=== Test metrics ===\nAccuracy@0.5={acc_05:.4f} | ROC-AUC={roc_auc:.4f} | AP={ap:.4f} | Brier={brier:.4f}")

# ROC
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate"); plt.title("ROC Curve")
plt.legend(loc="lower right"); plt.tight_layout()
roc_path = os.path.join(OUTPUT_DIR, "roc_curve.png"); plt.savefig(roc_path, dpi=150); plt.show()

# PR
plt.figure(figsize=(6,4))
plt.plot(recall, precision, label=f"AP = {ap:.3f}")
plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title("Precision–Recall Curve")
plt.legend(loc="lower left"); plt.tight_layout()
pr_path = os.path.join(OUTPUT_DIR, "pr_curve.png"); plt.savefig(pr_path, dpi=150); plt.show()

# Calibration
def reliability_curve(y_true, y_prob, n_bins=10):
    bins = np.linspace(0.0, 1.0, n_bins+1)
    idx = np.digitize(y_prob, bins) - 1
    frac_pos, mean_prob = [], []
    for b in range(n_bins):
        sel = (idx == b)
        if sel.sum() == 0:
            frac_pos.append(np.nan); mean_prob.append((bins[b]+bins[b+1])/2)
        else:
            frac_pos.append(y_true[sel].mean()); mean_prob.append(y_prob[sel].mean())
    return np.array(mean_prob), np.array(frac_pos)

mp, fp = reliability_curve(y_true, y_prob, n_bins=10)
plt.figure(figsize=(5,4))
plt.plot([0,1],[0,1],'--'); plt.plot(mp, fp, marker='o')
plt.xlabel("Mean predicted probability"); plt.ylabel("Fraction of positives")
plt.title("Calibration (Reliability) Curve")
plt.tight_layout()
cal_path = os.path.join(OUTPUT_DIR, "calibration.png"); plt.savefig(cal_path, dpi=150); plt.show()

# Tune F1 threshold + confusion matrix
candidates = np.linspace(0.1, 0.9, 81)
best_t, best_f1 = 0.5, -1.0
for t in candidates:
    hat = (y_prob >= t).astype(int)
    f1 = f1_score(y_true, hat, zero_division=0)
    if f1 > best_f1:
        best_f1, best_t = f1, float(t)

with open(os.path.join(OUTPUT_DIR, "threshold.json"), "w", encoding="utf-8") as f:
    json.dump({"best_threshold": best_t, "best_f1": float(best_f1)}, f, indent=2)

from sklearn.metrics import confusion_matrix
y_pred_best = (y_prob >= best_t).astype(int)
cm = confusion_matrix(y_true, y_pred_best)

plt.figure(figsize=(5.5,4.5))
if USE_SNS:
    sns.heatmap(cm, annot=True, fmt="d", cbar=False)
else:
    plt.imshow(cm, interpolation="nearest"); plt.colorbar()
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, str(cm[i, j]),
                     ha="center", va="center",
                     color="white" if cm[i, j] > cm.max()/2 else "black")
    plt.xticks([0,1], ["Incorrect(0)","Correct(1)"])
    plt.yticks([0,1], ["Incorrect(0)","Correct(1)"])
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title(f"Confusion Matrix (threshold={best_t:.2f})")
plt.tight_layout()
cm_path = os.path.join(OUTPUT_DIR, "confusion_matrix.png")
plt.savefig(cm_path, dpi=150); plt.show()

# Save metrics bundle
with open(os.path.join(OUTPUT_DIR, "metrics.json"), "w", encoding="utf-8") as f:
    json.dump({
        "accuracy_at_0.5": float(acc_05),
        "roc_auc": float(roc_auc),
        "average_precision": float(ap),
        "brier": float(brier),
        "best_threshold": best_t,
        "best_f1": float(best_f1)
    }, f, indent=2)

print(f"""
[INFO] Recovery complete (dtype fix applied).
- model.keras  -> {KERAS_PATH}
- roc_curve.png, pr_curve.png, calibration.png, confusion_matrix.png
- threshold.json, metrics.json
""")
```
